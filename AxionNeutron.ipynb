# Axion-neutron coupling limits vs axion mass

Axion mass-PQ scale relation:
\begin{equation}
m_{a}=5.70(7) \mu \mathrm{eV}\left(\frac{10^{12} \,\mathrm{GeV}}{f_{a}}\right)
\end{equation}

Axion-neutron coupling
\begin{equation}
g_{an} \equiv \frac{C_{an} m_{n}}{f_{a}} = 1.644\times 10^{-7} C_{an}\frac{m_a}{\mathrm{eV}}
\end{equation}

Model dependent constant:
\begin{equation}
C_{an}  = 
\begin{cases}
  -0.02(4) & {\rm KSVZ} \\
  [-0.16,0.26] & {\rm DFSZ\,I} \\
  [-0.16,0.26] & {\rm DFSZ\,II}
\end{cases}
\end{equation}
Like the axion-electron coupling the range of values for DFSZ I and II come from the perturbativity of the Yukawa couplings with sets the range $0.28<v_{u} / v_{d}<140$ for the Higgs vevs. I include the KSVZ line even though it is zero within errors.

```python
% matplotlib inline
from PlotFuncs import BlackHoleSpins, AxionNeutron, MySaveFig, FigSetup
import matplotlib.pyplot as plt

fig, ax = FigSetup(Shape='Rectangular', ylab='$|g_{an}|$', mathpazo=True,
                   g_min=1e-17, g_max=1e-2, m_min=1e-22, m_max=1e-2, FrequencyAxis=True, N_Hz=1)
AxionNeutron.QCDAxion(ax)
AxionNeutron.Haloscopes(ax, projection=False)
AxionNeutron.StellarBounds(ax)
AxionNeutron.LabExperiments(ax, projection=False)

BlackHoleSpins(ax, 0.93957, [3e-11, 2e-15], text_on=False)
plt.text(1.5e-12, 1e-14, r'{\bf Black}', fontsize=25, ha='center')
plt.text(1.5e-12, 1e-14 / 4, r'{\bf Hole}', fontsize=25, ha='center')
plt.text(1.5e-12, 1e-14 / 16, r'{\bf Spins}', fontsize=25, ha='center');

MySaveFig(fig, 'AxionNeutron')
```
