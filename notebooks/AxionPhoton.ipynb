# Axion-photon coupling limits vs axion mass

Axion mass-PQ scale relation:
\begin{equation}
m_{a}=5.70(7) \mu \mathrm{eV}\left(\frac{10^{12} \,\mathrm{GeV}}{f_{a}}\right)
\end{equation}

Axion-photon coupling
\begin{equation}
g_{a \gamma} \equiv \frac{\alpha}{2 \pi} \frac{C_{a\gamma}}{f_{a}}=2.0 \times 10^{-16} C_{a \gamma} \frac{m_{a}}{\mu \mathrm{eV}} \mathrm{GeV}^{-1}
\end{equation}

Model dependent constant:
\begin{equation}
C_{a\gamma} \equiv \frac{E}{N} - 1.92(4) = 
\begin{cases}
  -1.92, & {\rm KSVZ} \\
  0.75, & {\rm DFSZ\,I} \\
  -1.25, & {\rm DFSZ\,II}
\end{cases}
\end{equation}

```python
% matplotlib inline
from numpy import *
import matplotlib.pyplot as plt
from PlotFuncs import FigSetup, AxionPhoton, MySaveFig, BlackHoleSpins

fig, ax = FigSetup(Shape='Rectangular', ylab='$|g_{a\gamma}|$ [GeV$^{-1}$]', mathpazo=True)

AxionPhoton.QCDAxion(ax, C_center=abs(5 / 3 - 1.92) * (44 / 3 - 1.92) / 2, C_width=0.7, vmax=1.1)
AxionPhoton.Cosmology(ax)
AxionPhoton.StellarBounds(ax)
AxionPhoton.Haloscopes(ax)
AxionPhoton.Helioscopes(ax)
AxionPhoton.LSW(ax)
AxionPhoton.LowMassAstroBounds(ax)
AxionPhoton.ALPdecay(ax)
AxionPhoton.NeutronStars(ax)
BlackHoleSpins(ax, (1 / 137) * (1 / (2 * pi)), [2.1e-12, 1.5e-17], rotation=-90, alpha=0.3)

# this is bad, need a better solution to this
plt.arrow(0.265, 0.535, 0, 0.035, transform=fig.transFigure, figure=fig,
          length_includes_head=True, lw=1,
          head_width=0.007, head_length=0.016, overhang=0.13,
          edgecolor='crimson', facecolor='crimson', clip_on=True)

MySaveFig(fig, 'AxionPhoton')
```
