# Axion coupling to neutron electric dipole moment

QCD axion-edm coupling defined as
\begin{equation}
g_{an\gamma}^{\mathrm{QCD}} \approx \frac{2.4 \times 10^{-16}}{f_{a}} e \cdot \mathrm{cm} \approx 5.9 \times 10^{-10}\left(\frac{m_{a}}{\mathrm{eV}}\right) \mathrm{GeV}^{-2}
\end{equation}

```python
from numpy import *
import matplotlib.pyplot as plt
from PlotFuncs import BlackHoleSpins, AxionEDM, MySaveFig, FigSetup

fig, ax = FigSetup(Shape='Rectangular', ylab=r'$|g_{an\gamma}|$ [GeV$^{-2}$]', mathpazo=True,
                   g_min=1e-22, g_max=1e-2, m_min=1e-20, m_max=1e-2, FrequencyAxis=True, N_Hz=1)

AxionEDM.QCDAxion(ax)
AxionEDM.nEDM(ax)
AxionEDM.SN1987A(ax)
AxionEDM.CASPEr(ax)

BlackHoleSpins(ax, 6.4e-16 * 1e12, [0.7e-13, 0.2e-18], fs=23, rotation=90, text_col='k', text_on=True, alpha=1,
               facecolor='gray')
plt.gcf().text(0.475, 0.29, 'Stellar', ha='center', alpha=0.75)
plt.gcf().text(0.475, 0.26, 'BHs', ha='center', alpha=0.75)
plt.gcf().text(0.195, 0.15, 'SM \n BHs', alpha=0.75, ha='center')

MySaveFig(fig, 'AxionEDM')
```
